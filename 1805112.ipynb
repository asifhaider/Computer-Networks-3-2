```python
# import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_style('darkgrid')
sns.set_context('paper')
```

```python
input_file = open("1805112.txt", "r")
lines = input_file.readlines()
input_file.close()
```

```python
# creating output lists
throughput = []
delay = []
delivery = []
drop = []
area_size = ["250x250", "500x500", "750x750", "1000x1000", "1250x1250"]
node_count = [20, 40, 60, 80, 100]
flow_count = [10, 20, 30, 40, 50]
```

```python
# parsing the input file
for l in lines:
    if l.startswith("Network"):
        throughput.append(float(l.split(" ")[-2]))
    elif l.startswith("End"):
        delay.append(float(l.split(" ")[-2]))
    elif l.startswith("Packet delivery"):
        delivery.append(float(l.split(" ")[-2]))
    elif l.startswith("Packet drop"):
        drop.append(float(l.split(" ")[-2]))
```

```python
df1 = pd.DataFrame(zip(area_size, throughput))
df1.columns = ['Area Size', 'Network Throughput']
df1

df2 = pd.DataFrame(zip(area_size, delay))
df2.columns = ['Area Size', 'End to End Average Delay']
df2

df3 = pd.DataFrame(zip(area_size, delivery))
df3.columns = ['Area Size', 'Packet Delivery Ratio']
df3

df4 = pd.DataFrame(zip(area_size, drop))
df4.columns = ['Area Size', 'Packet Drop Ratio']
df4
```

```python
# plotting graphs
fig, axes = plt.subplots(figsize = (12,5))
sns.lineplot(data = df1, x='Area Size', y='Network Throughput',  marker='o')
plt.title('Network Throughput vs Area Size')
plt.xlabel('Area Size (m x m)') 
plt.ylabel('Network Throughput (bits/sec)')


fig, axes = plt.subplots(figsize = (12,5))
sns.lineplot(data = df2, x='Area Size', y='End to End Average Delay', marker='o')
plt.title('End to End Average Delay vs Area Size')
plt.xlabel('Area Size (m x m)') 
plt.ylabel('End to End Average Delay (sec)')


fig, axes = plt.subplots(figsize = (12,5))
sns.lineplot(data = df3, x='Area Size', y='Packet Delivery Ratio',  marker='o')
plt.title('Packet Delivery Ratio vs Area Size')
plt.xlabel('Area Size (m x m)') 
plt.ylabel('Packet Delivery Ratio (percent)')


fig, axes = plt.subplots(figsize = (12,5))
sns.lineplot(data = df4, x='Area Size', y='Packet Drop Ratio',  marker='o')
plt.title('Packet Drop Ratio vs Area Size')
plt.xlabel('Area Size (m x m)') 
plt.ylabel('Packet Drop Ratio (percent)')
```
